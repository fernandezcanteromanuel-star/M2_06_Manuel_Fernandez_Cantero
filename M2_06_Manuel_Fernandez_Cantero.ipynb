```python
#Radio del círculo
from math import pi


def area_circulo(radio):
    return pi*radio*radio

radio = float(input('Radio del círculo: '))

area_circulo(radio)
    
```

```python
#Mayor()

def lee_numero():
    n = int(input('Quién es Diego sniper?: '))
    i = 0
    lista = []
    while i < n:
        numero = float(input('Número: '))
        lista.append(numero)
        i += 1
    return lista

def mayor(lista):
    lista.sort()
    print(lista)
    return lista[-1]

a = lee_numero()

mayor(a)
```
